```python
#Detect outliers using percentile
from numpy import percentile
```

```python
q25,q75 = percentile(train_data.loc[:,'y'],25), percentile(train_data.loc[:,'y'],75)
iqr = q75-q25
```

```python
q25,q75
```

```python
iqr
```

```python
#calculate the cutoff outlier
cutoff = iqr*1.5
lower,upper = q25 - cutoff , q75+cutoff
```

```python
lower,upper
```

```python
#identify the outliers
outliers = [x for x in train_data.loc[:,'y'] if x < lower or x > upper]
```

```python
outliers
```

```python
print(f"Identfied the Outliers are {len(outliers)}")
```

```python
outliers_removed = [x for x in train_data.loc[:,'y'] if x >= lower or x <=upper]
```
